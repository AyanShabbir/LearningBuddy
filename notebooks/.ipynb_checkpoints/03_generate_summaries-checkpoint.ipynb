```python
# 03_generate_summaries.py or notebook cell
import pandas as pd
from tqdm import tqdm
from src.summary_generator import SummaryGenerator

# Load original dataset
df = pd.read_csv("../data/example_study_pack.csv")

# Initialize summary generator
sg = SummaryGenerator()

# Create columns for summaries
df['summary_simple'] = ""
df['summary_medium'] = ""
df['summary_advanced'] = ""

# Generate summaries with progress bar
for idx, row in tqdm(df.iterrows(), total=len(df), desc="Generating summaries"):
    content = row['content']
    df.at[idx, 'summary_simple'] = sg.summarize(content, level='simple')
    df.at[idx, 'summary_medium'] = sg.summarize(content, level='medium')
    df.at[idx, 'summary_advanced'] = sg.summarize(content, level='advanced')

# Save to CSV
df.to_csv("../data/example_study_pack_with_summaries.csv", index=False)

print("Saved example_study_pack_with_summaries.csv successfully!")
```

```python
-- BigQuery AI SQL: Generate summaries for lesson content
-- Replace `your-project-id` and `ai_learning_buddy` with your dataset

SELECT
  id,
  subject,
  level,
  text,
  AI.GENERATE_TEXT(
    MODEL `bqai.text-bison`,
    PROMPT => CONCAT(
      "Summarize the following text at three difficulty levels:\n",
      "Text: ", text, "\n",
      "Format: simple summary | medium summary | advanced summary"
    )
  ) AS generated_summaries
FROM `your-project-id.ai_learning_buddy.lesson_content`
LIMIT 10;
```
